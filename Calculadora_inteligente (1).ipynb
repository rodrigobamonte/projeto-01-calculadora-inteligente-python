# AGORA DAREMOS INÍCIO AOS CÓDIGOS PARA PODER MONTAR A CALCULADORA INTELIGENTE

```python
continuar = True

while continuar:

    print("Bem-vindo à Calculadora Inteligente!")

    print("1 - Operações matemáticas")
    print("2 - Calcular média")
    print("3 - Comparar números")
    print("4 - Sair")

    opcao_menu = input("Escolha uma opção: ")

    if opcao_menu == "1":

        print("Você escolheu Operações Matemáticas.")

        print("1 - Soma")
        print("2 - Subtração")
        print("3 - Multiplicação")
        print("4 - Divisão")
        print("5 - Porcentagem")
        print("6 - Potência")
        print("7 - Raiz Quadrada")

        opcao_operacao = input("Escolha uma operação: ")

        # SOMA
        if opcao_operacao == "1":

            numero_1 = float(input("Digite o primeiro número: "))
            numero_2 = float(input("Digite o segundo número: "))

            resultado = numero_1 + numero_2

            print("Resultado:", resultado)

            if resultado > 0:
                print("O resultado é positivo.")

            elif resultado < 0:
                print("O resultado é negativo.")

            else:
                print("O resultado é igual a zero.")


        # SUBTRAÇÃO
        elif opcao_operacao == "2":

            numero_1 = float(input("Digite o primeiro número: "))
            numero_2 = float(input("Digite o segundo número: "))

            resultado = numero_1 - numero_2

            print("Resultado:", resultado)

            if resultado > 0:
                print("O resultado é positivo.")

            elif resultado < 0:
                print("O resultado é negativo.")

            else:
                print("O resultado é igual a zero.")


        # MULTIPLICAÇÃO
        elif opcao_operacao == "3":

            numero_1 = float(input("Digite o primeiro número: "))
            numero_2 = float(input("Digite o segundo número: "))

            resultado = numero_1 * numero_2

            print("Resultado:", resultado)

            if resultado > 0:
                print("O resultado é positivo.")

            elif resultado < 0:
                print("O resultado é negativo.")

            else:
                print("O resultado é igual a zero.")


        # DIVISÃO
        elif opcao_operacao == "4":

            numero_1 = float(input("Digite o primeiro número: "))
            numero_2 = float(input("Digite o segundo número: "))

            if numero_2 == 0:

                print("Não é possível dividir por zero.")

            else:

                resultado = numero_1 / numero_2

                print("Resultado:", resultado)

                if resultado > 0:
                    print("O resultado é positivo.")

                elif resultado < 0:
                    print("O resultado é negativo.")

                else:
                    print("O resultado é igual a zero.")


        # PORCENTAGEM
        elif opcao_operacao == "5":

            valor = float(input("Digite o valor: "))
            porcentagem = float(input("Digite a porcentagem: "))

            resultado = valor * porcentagem / 100

            print("Resultado:", resultado)

            if resultado > 0:
                print("O resultado é positivo.")

            elif resultado < 0:
                print("O resultado é negativo.")

            else:
                print("O resultado é igual a zero.")


        # POTÊNCIA
        elif opcao_operacao == "6":

            base = float(input("Digite a base: "))
            expoente = float(input("Digite o expoente: "))

            resultado = base ** expoente

            print("Resultado:", resultado)

            if resultado > 0:
                print("O resultado é positivo.")

            elif resultado < 0:
                print("O resultado é negativo.")

            else:
                print("O resultado é igual a zero.")


        # RAIZ QUADRADA
        elif opcao_operacao == "7":

            numero = float(input("Digite o número da raiz: "))

            if numero < 0:

                print("Operação inválida. Não é possível calcular raiz quadrada de número negativo.")

            else:

                resultado = numero ** 0.5

                print("Resultado:", resultado)

                if resultado > 0:
                    print("O resultado é positivo.")

                elif resultado < 0:
                    print("O resultado é negativo.")

                else:
                    print("O resultado é igual a zero.")


        else:

            print("Operação inválida.")


    elif opcao_menu == "2":

        print("Você escolheu Calcular Média.")


    elif opcao_menu == "3":

        print("Você escolheu Comparar Números.")


    elif opcao_menu == "4":

        print("Encerrando a Calculadora...")

        continuar = False


    else:

        print("Opção inválida! Por favor, escolha uma opção entre 1 e 4.")

```
